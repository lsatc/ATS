```python
import fitz  # PyMuPDF library
from deep_translator import GoogleTranslator
from langdetect import detect
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download("stopwords")
nltk.download('punkt')

# Initialize the translator
translator = GoogleTranslator()


def summerize(word_tokens):

  stop_words = set(stopwords.words('english'))

  filtered_sentence = [w for w in word_tokens if not w in stop_words]

  filtered_text = ' '.join(filtered_sentence)

  return filtered_text

# Extract text from every page in the PDF
def pdf_reader(pdf_file):
  text = ''
  for page_index in range(len(pdf_file)):
      page = pdf_file[page_index]
      text += page.get_text()
  pdf_file.close()
  return text

def clean(text):
  # Detect the language of the text
  detected_lang = detect(text)
  words = text.split()
  summerized_text = summerize(words)
  return summerized_text
  # If the text is in Persian, translate it to English
"""  if detected_lang == 'fa':
      translated_text = translator.translate(text, source='fa', target='en')
      words = translated_text.split()
      summerized_text = summerize(words)
      return summerized_text"""


def main():
  pdf = fitz.open('sample.pdf')
  text = pdf_reader(pdf)
  cleaned_text = clean(text)
  print(cleaned_text)


if __name__ == "__main__":
    main()
```
